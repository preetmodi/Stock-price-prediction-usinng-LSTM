```python
company_name = 'AMZN'
end = datetime.datetime.now()

start = datetime.datetime(end.year-10, end.month, end.day)
data = DataReader(
            company_name,
            'yahoo',
            start,
            end)
```

```python
# data['Date'] = pd.to_datetime(data.index, format='%y-%m-%d', errors='ignore')
data.info()
```

```python
plt.figure(figsize=(16,8))
plt.plot(data['Close'], label='Close Price history')
```

```python
dataset = pd.DataFrame(index=(range(0,len(data))),columns=['Date','Close'])
data = data.sort_index(ascending=True, axis=0)
dataset['Date'] = data.index
dataset = dataset.set_index(['Date'])
dataset['Close'] = data['Close']
# data['Date']
```

```python
split = int(len(dataset)*0.75)
X = dataset.values
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(X)
```

```python
train = scaled_data[:split,:]
valid = scaled_data[split:,:]
X_train = []
y_train = []
for i in range(60, len(train)):
    X_train.append(train[i-60:i, 0])
    y_train.append(train[i, 0])
    
X_train = np.array(X_train)
y_train = np.array(y_train)

X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```

```python
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(X_train.shape[1],1)))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))

model.compile(optimizer= 'adam',
             loss = 'mean_squared_error')
```

```python
model.fit(X_train, 
          y_train,
         epochs=2)
```

```python
test_data  = scaled_data[split-60:,:]
X_test = []
for i in range(60,(len(test_data))):
    X_test.append(test_data[i-60:i,:])
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0],X_test.shape[1],1))
y_test = X[split:,:]
result = model.predict(X_test)
result = scaler.inverse_transform(result)

rmse = np.sqrt(np.mean((y_test-result)**2))
```

```python
test = pd.DataFrame(index=dataset[split:].index, columns=['Close'])
# test['Date'] = dataset[split:].index
test['Close'] = result
# test = test.set_index(['Date'])
```

```python
plt.figure(figsize=(20,10))
plt.plot(dataset[:split], label='Training Data', color='blue')
plt.plot(dataset[split:], label='Test Data', color='blue')
plt.plot(test['Close'], label='Predicted Values', color='red')
plt.savefig(company_name+"_full_timeline.jpg", dpi=300, bbox_inches='tight')
plt.legend()
print("RMS error for this dataset is: ", rmse)
print()
print("Below shown is the graph suggesting the predicted values vs Actual values of the closing price of Tesla stock for the respective dates")
```

```python
X_validate = []
resultset = []
prediction_days = 10
X_validate.append(test_data[-60:,:])
X_validate = np.array(X_validate)
X_validate = np.reshape(X_validate, (X_validate.shape[0],X_validate.shape[1],1))
for i in range(prediction_days):
    predicted = model.predict(X_validate)
    resultset.append(predicted[0])
    X_validate = np.delete(X_validate, 0, axis=1)
    X_validate = np.reshape(np.append(X_validate, predicted), (1,60,1))
    
len(resultset)
```

```python
resultset= np.array(resultset)
resultset = scaler.inverse_transform(resultset)
```

```python
start = test.index[-1] + datetime.timedelta(days=1)
index = [start]
for i in range(prediction_days-1):
    start+=datetime.timedelta(days=1)
    index.append(start)
finalprediction = pd.DataFrame(index= index, columns=['Close'])
finalprediction['Close'] = resultset
```

```python
plt.figure(figsize=(16,5))
plt.plot(finalprediction, label = 'Predicted stock price for next '+ str(prediction_days) + ' Days', color='red')
plt.plot(test['Close'][-60:], label='Actual data for last 60 days', color='Blue')
plt.legend()
plt.savefig(company_name+"future.jpg", dpi=300, bbox_inches='tight')
plt.title('Stock price Prediction')
```
